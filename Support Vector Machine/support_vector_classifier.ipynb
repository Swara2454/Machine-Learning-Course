```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_redundant=0,n_clusters_per_class=2)
```

```python
X
```

```python
y
```

```python
df=pd.DataFrame(X,columns=['feature1','feature2'])
df['target']=y
```

```python
sns.scatterplot(data=df,x='feature1',y='feature2',hue='target')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
y_pred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

## To improve the accuracy we use different kernals

## RBF kernal

```python
from sklearn.svm import SVC
rbf=SVC(kernel='rbf')
rbf.fit(X_train,y_train)
y_pred1=rbf.predict(X_test)

print(classification_report(y_pred1,y_test))
print(confusion_matrix(y_pred1,y_test))
```

Thus by using RBF kernal the accurecy has increased to 90%

## Polynomial Kernal

```python
from sklearn.svm import SVC
poly=SVC(kernel='poly')
poly.fit(X_train,y_train)
y_pred2=poly.predict(X_test)

print(classification_report(y_pred2,y_test))
print(confusion_matrix(y_pred2,y_test))
```

Thus by using the polynomial kernal the accuracy is 88 which is less than rbf kernal

```python
from sklearn.svm import SVC
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
y_pred3=sigmoid.predict(X_test)

print(classification_report(y_pred3,y_test))
print(confusion_matrix(y_pred3,y_test))
```

Thus by using the sigmoid kernal the accuracy is 83 which is less than rbf and polynoinal kernal

## Hyperparameter Tunning with SVC

```python
from sklearn.model_selection import GridSearchCV

param_grid={
    'C':[0.1,1,10,100,1000],
    'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf']
}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred4=grid.predict(X_test)

print(classification_report(y_pred4,y_test))
print(confusion_matrix(y_pred4,y_test))
```

## By hyperparameter tunning we can increase the accuracy
