```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=6*np.random.rand(100,1)-3       ## Data points are created randomly from the range -3 to 3
y=0.5* X**2 +1.5*X + 2 + np.random.randn(100,1)        ## Data pionts for y are created in the format 1.5 x^2 + 0.5 x + 3 + Noice
```

```python
plt.scatter(X,y,color='r')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
## Trying linear regression

from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
```

```python
regression_1.fit(X_train,y_train)
```

```python
y_pred=regression_1.predict(X_test)
```

```python
plt.plot(X_train,regression_1.predict(X_train),color='r')
plt.scatter(X_train,y_train)
plt.xlabel('X label')
plt.ylabel("Y_label")
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

As the accuracy in very low through linearregression best fit line we use polynomial regression 

```python
## applying polynomial regression

from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
```

```python
X_poly_train=poly.fit_transform(X_train)
X_poly_test=poly.transform(X_test)
```

```python
X_poly_train
```

```python
X_poly_test
```

```python
regression=LinearRegression()
regression.fit(X_poly_train,y_train)
y_pred_new=regression.predict(X_poly_test)
y_pred_new
```

```python
r2_new=r2_score(y_test,y_pred_new)
r2_new
```

```python
plt.scatter(X_train,regression.predict(X_poly_train))   
plt.scatter(X_train,y_train)
```

Pipeline 

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    X_label=np.linspace(-3,3,200).reshape(200,1)
    poly_regression=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression()
    pipeline=Pipeline([
        ("poly_regression",poly_regression),
        ("lin_reg",lin_reg)
    ])

    pipeline.fit(X_train,y_train)
    y_pred_new=pipeline.predict(X_label)
    #plotting prediction line
    plt.plot(X_label, y_pred_new,'r', label="Degree " + str(degree), linewidth=2)
    plt.plot(X_train, y_train, "b.", linewidth=3)
    plt.plot(X_test, y_test, "g.", linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4, 0, 10])
    plt.show()
```

```python
poly_regression(6)
```
