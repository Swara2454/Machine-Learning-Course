```python
from sklearn.datasets import make_classification
```

## Without Smote

```python
## Making Dataset using sklearn

X,y=make_classification(
    n_samples=1000,
    n_redundant=0,
    n_features=2,
    n_clusters_per_class=1,
    weights=[0.90],
    random_state=12
)
```

```python
## Making dataframe

import pandas as pd
df1=pd.DataFrame(X,columns=['feature1','feature2'])
df2=pd.DataFrame(y,columns=['Target'])
final_df=pd.concat([df1,df2],axis=1)
```

```python
final_df.head()
```

```python
final_df['Target'].value_counts()
```

```python
## Upsampling 

import matplotlib.pyplot as plt
plt.scatter(final_df['feature1'],final_df['feature2'],c=final_df['Target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversampling=SMOTE()
X,y=oversampling.fit_resample(final_df[['feature1','feature2']],final_df['Target'])
```
