## Imputation techniques

Mean Technique --- Used when the data in normally distributed


```python
df['age_mean']=df['age'].fillna(df['age'].mean())
```

```python
df[['age','age_mean']]
```

Median technique--- Used when data contains outliers of the data is right or left scewed

```python
df['age_median']=df['age'].fillna(df['age'].median())
```

```python
df[['age','age_median','age_mean']]
```

Mode technique ---- Categorical values

```python
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
```

```python
mode_value=df[df['embarked'].notna()]['embarked'].mode()[0]
```

```python
df['embarked_mode']=df['embarked'].fillna(mode_value)
```

```python
df[['embarked_mode','embarked']]
```

```python
df['embarked_mode'].isnull().sum()
```
