```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel('flight_price.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
## Date split in day,month and year and drop the date_of_journey column

df['Journey_Day']=df['Date_of_Journey'].str.split('/').str[0]
df['Journey_Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Journey_Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
## Convert the datatype of dates into int 

df['Journey_Day']=df['Journey_Day'].astype(int)
df['Journey_Month']=df['Journey_Month'].astype(int)
df['Journey_Year']=df['Journey_Year'].astype(int)
```

```python
df.info()
```

```python
df['Dep_Time']
```

```python
## COnverting Dep_Time in Hours and Minutes

df['Dep_Time_Hrs']=df['Dep_Time'].str.split(':').str[0]
df['Dep_Time_Mins']=df['Dep_Time'].str.split(':').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
## Converted Dep_time_mins ans hrs to int
df[['Dep_Time_Mins','Dep_Time_Hrs']]=df[['Dep_Time_Mins','Dep_Time_Hrs']].astype(int)
```

```python
df.info()
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Hrs']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0]).str.split(':').str[0]
df['Arrival_Min']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0]).str.split(':').str[1]
```

```python
df.head()
```

```python
df.info()
```

```python
df[['Arrival_Hrs','Arrival_Min']]=df[['Arrival_Hrs','Arrival_Min']].astype(int)
```

```python
df.info()
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, np.nan:1, '4 stops':4})
```

```python
df.head()
```

```python
df['Duration_hrs']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0].fillna(0)
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
## Converting categorical values into numerical values

from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```
